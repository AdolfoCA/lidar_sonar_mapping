```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
# data = pd.read_csv('merged10.mcap.csv')
data = pd.read_csv('/media/aimas/aee54fa2-89ee-4a49-8b37-3d5c29135b72/cph_lidar_data/mcap/fletcher_cut/fletcher_cut_0.db3.mcap.csv')
data["Timestamp"] = pd.to_datetime(data["Timestamp"])
print(data)
```

```python
mean = data["Norm"].mean()
std = data["Norm"].std()

print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
x = data["Timestamp"].values
y = data["Norm"].values

plt.plot(x, y, label='Norm of IMU linear acceleration vector')
plt.plot(x, [mean + std] * len(x), '--', color='#FF0000', label='Standard Deviation')
plt.plot(x, [mean - std] * len(x), '--', color='#FF0000')
plt.xticks(rotation=30)
plt.gca().xaxis.set_major_formatter(mpl.dates.DateFormatter('%H:%M:%S'))
plt.grid(True)
plt.legend()
plt.show()
```
